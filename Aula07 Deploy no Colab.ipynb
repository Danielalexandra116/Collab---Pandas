```python
%%writefile AC02.py
import pandas as pd
import numpy as np
import streamlit as st

st.title('AC02 - Análises Base Covid')

df = pd.read_csv('covid.csv')

paises = []

paises.append('Todos')

for x in df['Country'].unique():
    paises.append(x)

lista_pais = st.sidebar.selectbox(
        'Selecione o país',
        (paises)
        )

col1, col2, col3 = st.columns(3)

if lista_pais == 'Todos':

    pais_agrupado = df.groupby('Country')

    col1.subheader('Total de casos')
    col1.subheader(df['total_cases'].sum())

    col2.subheader('Total de Mortes')
    col2.subheader(df['total_deaths'].sum())

    col3.subheader('Total de Testes')
    col3.subheader(df['total_tests'].sum())

    st.dataframe(df)

    #st.write
    st.line_chart(data=pais_agrupado[['total_deaths', 'total_cases']].sum())

else:
    df_filtrado = df[df['Country'] == lista_pais]

    data_agrupada = df_filtrado.groupby('date')

    col1.subheader('Tota de casos')
    col1.subheader(df_filtrado['total_cases'].sum())

    col2.subheader('Total de Mortes')
    col2.subheader(df_filtrado['total_deaths'].sum())

    col3.subheader('Total de Testes')
    col3.subheader(df_filtrado['total_tests'].sum())


    st.dataframe(df_filtrado)
    st.write(f'Total de Mortes no país {lista_pais} por dia:')
    st.line_chart(data_agrupada['total_deaths'].sum())

    st.write(f'Total de Novos casos no país {lista_pais} por dia:')
    st.line_chart(data_agrupada['new_cases'].sum())

    st.write(f'Total de testes realizados no país {lista_pais} por dia:')
    st.line_chart(data_agrupada['total_tests'].sum())
```

```python
%%writefile aplicacao.py
import streamlit as st
import plotly.graph_objs as go
import pandas as pd

st.text('Dados do Covid')

DATA_URL = ('https://raw.githubusercontent.com/nikkisharma536/streamlit_app/master/covid.csv')

data = pd.read_csv(DATA_URL)
data = data.rename(columns={'Latitude': 'latitude', 'Longitude': 'longitude'})

df2 = data[['latitude', 'longitude', 'total_cases']]

df = data

df = df[df['Country'] == 'Argentina']

linha_grafico_1 = go.Scatter(x=df['date'].values, y=df['new_cases'].values, text='Novos')
linha_grafico_2 = go.Scatter(x=df['date'].values, y=df['total_deaths'].values, text='Mortos')

linhas_grafico = [linha_grafico_1, linha_grafico_2]
	
st.write(linhas_grafico)
st.map(df2)
```
