EXERCÍCIO PARA FAZER WEB SCRAPING DO SITE DA GLOBO E BUSCAR AS NOTÍCIAS MAIS VISTAS - TOP RANKING

```python
# Biblioteca para Web Scraping
from bs4 import BeautifulSoup
```

```python
# Biblioteca para fazer Requests da Web
import requests
```

```python
# Fazendo um request da globo e gravando em uma variável
response = requests.get('https://www.globo.com/')
```

```python
# Analisando o resultado da request - html
response.text
```

```python
# Transformando com o parser para html
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Analisando todos os 'h3' da página web (após inspecionar a página - 1ª busca)
soup.find_all('h3')
```

```python
# Analisando o index para separação de dados
top_not = soup.find_all('h3')
for x in enumerate(top_not):
  print(x)
```

```python
# Buscando todos os h3 da página que estão contidos na classe post_title (outra maneira)
soup.find_all('h3', class_='post__title')
```

```python
# Separando por índice só as notícias relativas à pesquisa
top = top_not[44:61]
```

```python
# Analisando o resultado da separação
top
```

```python
# Lista vazia para guardar as notícias
lista_not = []
```

```python
# Guardando as notícias em formato de texto na lista criada
for i in top:
  lista_not.append(i.text)
  print(i.text)
```

```python
# Conferindo a lista final com as notícias da pesquisa
lista_not
```
